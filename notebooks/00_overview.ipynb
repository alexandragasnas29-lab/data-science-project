## What is python

[Python](http://www.python.org/) is a modern, general-purpose, object-oriented, high-level programming language.

General characteristics of Python:

* **simple and clean:** Python code is easy-to-read and intuitive, has easy-to-learn minimalistic syntax, and maintainability scales well with size of projects.
* **expressive language:** Fewer lines of code, fewer bugs, easier to maintain.

Technical details:

* **dynamically typed:** No need to define the type of variables, function arguments or return types.
* **automatic memory management:** No need to explicitly allocate and deallocate memory for variables and data arrays. No memory leak bugs. 
* **interpreted:** No need to compile the code. The Python interpreter reads and executes the python code directly.

Advantages:

* The main advantage is ease of programming, minimizing the time required to develop, debug and maintain the code.
* Well designed language that encourage many good programming practices:
 * Modular and object-oriented programming, good system for packaging and re-use of code. This often results in more transparent, maintainable and bug-free code.
 * Documentation tightly integrated with the code.
* A large standard library, and a large collection of add-on packages.

Disadvantages:

* Since Python is an interpreted and dynamically typed programming language, the execution of python code can be slow compared to compiled statically typed programming languages, such as C and Fortran. 
* Somewhat decentralized, with different environment, packages and documentation spread out at different places. Can make it harder to get started.

## Mutability vs Immutability

In Python, objects can be either **mutable** (changeable) or **immutable** (unchangeable after creation). This is a fundamental concept that affects how you work with different data types.

### Immutable Types

**Cannot be changed** after creation. Any operation that appears to modify them actually creates a new object.

Common immutable types:
- `int`, `float` — numbers
- `str` — strings
- `tuple` — ordered collection (covered later)
- `bool` — True/False

````python
# Strings are immutable
text = "hello"
print(f"Original: {text}")

# This creates a NEW string, doesn't change the original
new_text = text.upper()
print(f"After .upper(): original is still '{text}', new is '{new_text}'")
````

### Mutable Types

**Can be changed** in-place after creation without creating a new object.

Common mutable types:
- `list` — ordered collection
- `dict` — key-value pairs
- `set` — unordered unique values

````python
# Lists are mutable
numbers = [1, 2, 3]
print(f"Original: {numbers}")

# This modifies the SAME list object
numbers.append(4)
print(f"After .append(4): {numbers}")

# You can also change elements directly
numbers[0] = 99
print(f"After numbers[0] = 99: {numbers}")
````

### Why This Matters

Understanding mutability helps you:
- **Predict behavior**: Know whether an operation changes the original or creates something new
- **Avoid bugs**: Especially when multiple variables refer to the same mutable object
- **Write efficient code**: Mutable types can be modified in-place without creating copies

You'll see this concept throughout the course, especially in the notebooks on [strings](basic/01_strings.ipynb), [lists](basic/03_lists.ipynb), and [dictionaries](basic/04_dictionaries.ipynb).

## Functions vs Methods

In Python you will encounter two ways of "doing things" with data:

- **Functions** are called *on their own* and take input as arguments: `function(object)`
- **Methods** are called *on an object* using dot notation: `object.method()`

**Key difference:** A method is a function that *belongs to* a specific type of object.

| Syntax | Example | What it does |
|--------|---------|--------------|
| `function(object)` | `len("hello")` | Counts characters — works on strings, lists, etc. |
| `object.method()` | `"hello".upper()` | Converts to uppercase — only works on strings |

A simple way to think about it:
- **Functions** are general tools (like `print()`, `len()`, `type()`)
- **Methods** are specific skills that an object *knows how to do* (like `.upper()`, `.replace()`, `.append()`)

````python
name = "Python" # a string variable

# Function — len() is a built-in function, works on many types
print(len(name))

# Method — .upper() belongs to strings
print(name.upper())

# Method — .replace() also belongs to strings
print(name.replace("P", "J"))

# You can even chain methods together
print("  hello world  ".strip().title())
````

````python
# Use dir() to see all methods available on an object
# (ignore the ones starting with __ for now)
public_methods = [m for m in dir(name) if not m.startswith('_')]
print(public_methods)
````

## Whitespace Formatting

Many languages use curly braces to delimit blocks of code. Python uses indentation (four spaces):

````python
for i in [1, 2]:
    print(i)                    # first line in "for i" block
    for j in [3, 4]:
        print(j)                # first line in "for j" block
        print(i + j)            # last line in "for j" block
    print(i)                    # last line in "for i" block
print("done looping")
````
