# [`for` loops](https://docs.python.org/3/tutorial/controlflow.html#for-statements)

## Looping lists

```python
my_list = [1, 2, 3, 4, 'Python', 'is', 'neat']
for item in my_list:
    print(item)
```

### `break`
Stop the execution of the loop.

```python
for item in my_list:
    if item == 'Python':
        break
    print(item)
```

### `continue`
Continue to the next item without executing the lines occuring after `continue` inside the loop.

```python
for item in my_list:
    if item == 4:
        continue
    print(item)
```

### `enumerate()`
In case you need to also know the index:

```python
for idx, val in enumerate(my_list):
    print('idx: {}, value: {}'.format(idx, val))
```






## Looping dictionaries

```python
my_dict = {'hacker': True, 'age': 72, 'name': 'John Doe'}
for key in my_dict:
    print(key)
```

```python
for key, val in my_dict.items():
    print('{}={}'.format(key, val))
```

## `range()`

```python
for number in range(5):
    print(number)
```

```python
for number in range(2, 5): # start at 2 inclusive, end at 5 excluding it 
    print(number)
```

```python
for number in range(0, 10, 2):  # last one is step
    print(number)
```

## `zip()` - Looping Over Multiple Iterables

The `zip()` function allows you to iterate over multiple sequences in parallel. It creates pairs (or tuples) from elements at corresponding positions.

### Basic `zip()` Examples

```python
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
cities = ['Copenhagen', 'Stockholm', 'Oslo']

# Loop over two lists simultaneously
for name, age in zip(names, ages):
    print(f'{name} is {age} years old')
```

```python
# Loop over three lists simultaneously
for name, age, city in zip(names, ages, cities):
    print(f'{name} ({age}) lives in {city}')
```

### Creating Dictionaries with `zip()`

```python
# Create a dictionary from two lists
keys = ['name', 'age', 'occupation']
values = ['Diana', 28, 'Engineer']

person = dict(zip(keys, values))
print('Person dict:', person)

# Alternative using dictionary comprehension
person2 = {k: v for k, v in zip(keys, values)}
print('Person dict 2:', person2)
```

### Unequal Length Lists

**Important:** `zip()` stops when the shortest iterable is exhausted.

```python
list1 = [1, 2, 3, 4, 5]
list2 = ['a', 'b', 'c']

# Only pairs until the shortest list
for num, letter in zip(list1, list2):
    print(f'{num}: {letter}')
# Output: 1:a, 2:b, 3:c (4 and 5 are ignored)
```

### Practical Examples

```python
# Calculate total price
items = ['apple', 'banana', 'orange']
prices = [10, 5, 8]
quantities = [3, 5, 2]

total = 0
for item, price, qty in zip(items, prices, quantities):
    cost = price * qty
    print(f'{item}: {qty} x {price} = {cost}')
    total += cost
print(f'Total: {total}')
```

```python
# Compare two lists
expected = [1, 2, 3, 4, 5]
actual = [1, 2, 4, 4, 5]

for i, (exp, act) in enumerate(zip(expected, actual)):
    if exp != act:
        print(f'Mismatch at index {i}: expected {exp}, got {act}')
```

### Converting `zip()` to Other Data Structures

```python
# zip() returns an iterator, convert to list to see all pairs
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']

paired = list(zip(list1, list2))
print('Paired:', paired)  # [(1, 'a'), (2, 'b'), (3, 'c')]

# Unzip: convert back to separate lists
numbers, letters = zip(*paired)
print('Numbers:', numbers)
print('Letters:', letters)
```
