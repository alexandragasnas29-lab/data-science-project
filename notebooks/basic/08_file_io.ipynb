Reading and writing files.

## Working with paths

```python
import os

current_file = os.path.realpath('07_file_io.ipynb')  #path.realpath is a method on a method, it will give you the absolute path of the file, it will resolve any symbolic links and return the canonical path of the file.
print('current file: {}'.format(current_file))
# Note: in .py files you can get the path of current file by __file__
```

```python
current_dir = os.path.dirname(current_file)  
print('current directory: {}'.format(current_dir))
# Note: in .py files you can get the dir of current file by os.path.dirname(__file__)
```

```python
data_dir = os.path.join(os.path.dirname(current_file), 'data')
print('data directory: {}'.format(data_dir))
```

```python
os.getwd()
os.chdir(data_dir)
print('current directory: {}'.format(os.getcwd()))  
```

### Checking if path exists

```python
print('exists: {}'.format(os.path.exists(current_dir)))
print('is file: {}'.format(os.path.isfile(current_file)))
print('is directory: {}'.format(os.path.isdir(data_dir)))
```

## Reading files

```python
file_path = os.path.join(data_dir, 'simple_file.txt')

with open(file_path, 'r') as simple_file: #'r' is for read mode, 'w' is for write mode, 'a' is for append mode
    for line in simple_file:
        print(line.strip())
#this opens and closes the connection to the file after the lines are executed        
```

## Writing files

```python
new_file_path = os.path.join(data_dir, 'new_file.txt')

with open(new_file_path, 'w') as my_file: #this will create a new file called new_file.txt in the data directory, if the file already exists it will overwrite it, if the file does not exist it will create it
    my_file.write('This is my first file that I wrote with Python.')
```

Now go and check that there is a new_file.txt in the data directory. After that you can delete the file by:

```python
if os.path.exists(new_file_path):  # make sure it's there
    os.remove(new_file_path)
```
